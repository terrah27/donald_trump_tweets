```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from plotly import graph_objs as go

import spacy
```

```python
# setting up default plotting parameters
%matplotlib inline

plt.rcParams['figure.figsize'] = [20.0, 7.0]
plt.rcParams.update({'font.size': 22,})

sns.set_palette('viridis')
sns.set_style('white')
sns.set_context('talk', font_scale=0.8)
```

```python
df = pd.read_json('condensed_2018.json')
df.head()
```

```python
df.shape
```

```python
fig = go.Figure(
    data=[go.Bar(x=[1, 2, 3], y=[1, 3, 2])],
    layout=go.Layout(
        title=go.layout.Title(text="A Bar Chart")
    )
)
#fig.show()
```

```python
df['week'] = df['created_at'].dt.week
df.head()
```

```python
df.tail()
```

```python
g = sns.countplot(df.week, palette='viridis')
#g.set_xticklabels([])
sns.despine(left=True, bottom=True)
plt.xlabel('Week')
plt.ylabel('Number of Tweets')
plt.title('Trump Tweets by Week', fontsize=30)
plt.tick_params(axis='x', which='major', labelsize=10)
plt.show();
```

```python
df['day_of_week'] = df['created_at'].dt.day_name()
df.head()
```

```python
g = sns.countplot(df.day_of_week, palette='viridis')
#g.set_xticklabels([])
sns.despine(left=True, bottom=True)
plt.xlabel('Day of Week')
plt.ylabel('Number of Tweets')
plt.title('Trump Tweets by Day of Week', fontsize=30)
plt.tick_params(axis='x', which='major', labelsize=10)
plt.show();
```

```python
g = sns.countplot(df.source)
#g.set_xticklabels([])
sns.despine(left=True, bottom=True)
plt.xlabel('Source')
plt.ylabel('Number of Tweets')
plt.title('Trump Tweets by Source', fontsize=30)
plt.tick_params(axis='x', which='major', labelsize=10)
plt.show();
```
